```python
import pyupbit
import numpy as np
import datetime
import cufflinks as cf
import pandas as pd
```

```python
ddf = pyupbit.get_ohlcv("KRW-BTC", count=200)
price = {'Buy':[], 'Sell':[]}
seed = 100000
num = 0
bday= pd.DataFrame()
sday= pd.DataFrame()
scale = 1000
```

```python
def buy(today, yesterday):
    global buy_price, price, seed, num, bday, scale
    rise1 = (yesterday.iloc[0]['close'] - yesterday.iloc[0]['open'])/yesterday.iloc[0]['open']
    rise2 = (today.iloc[0]['close'] - today.iloc[0]['open'])/today.iloc[0]['open']
    if rise1 > 0 and rise2 >= rise1 * 0.5 and seed >= 50000:
        bday = pd.concat([bday, today])
        buy_price = list(today['open'])[0]*(1+rise2)/scale
        print('''
    {}일 {}원 매수
              '''.format(list(today.index)[0], buy_price))
        price['Buy'].append(buy_price)
        seed -= buy_price
        num += 1
```

```python
def sell(today):
    global sell_price, price, seed, sday, num, scale
    sday = pd.concat([sday, today])
    sell_price = list(today['open'])[0]/scale
    print('''
    {}일 {}원 매도
              '''.format(list(today.index)[0], sell_price))
    price['Sell'].append(sell_price)
    seed += sell_price
    num -= 1
```

```python
for d in list(ddf.index):
    try :
        if num == 0:
            buy(ddf.loc[[d]], ddf.loc[[yesterday]])
        elif num == 1 :
            sell(ddf.loc[[d]])
            
    except NameError:
        print('시작')
    yesterday = d
    

print('총 수익: {}'.format(sum(price['Sell'])-sum(price['Buy'])))
print(len(bday), len(sday))
print('단순수익률: {} %'.format((ddf.iloc[199,3]-ddf.iloc[0,3])/ddf.iloc[0,3] * 100))
```

```python
bday['Buy'] = price['Buy']
sday['Sell'] = price['Sell']
```

```python
cf.go_offline(connected=True)
qf=cf.QuantFig(ddf,title='BTC',legend='top',name='Buy',
              up_color='red',
              down_color='blue')
qf.iplot()
```
